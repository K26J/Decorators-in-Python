# Decorators:

Decorator in a python is a function that recives another function as an input and add some functionality(Decoration) to it and returns it.

There are two types of decorators in python
* Builtin like `@Staticmethod` `@Classmethod` `Abstractmethod` etc.
* User Defined

```python
# Let's define a decorator which adds decoration at begining and at te end of the code as:
# Execution Started....
#....Execution Completed

def execution(func):
  def wrapper():
    print("Execution Started....")
    func()
    print("....Execution Completed")
  return wrapper


def hello():
  print("Ky re kay challay")


a= execution(hello)
a()
```

```python
# Smart way to do this using @
def execution(func):
  def wrapper():
    print("."*1000)
    func()
    print("."*1000)
  return wrapper


@execution
def myname():
  print('Keshav Mahesh Joshi')


myname()
```

```python
# Let's make a decorator runner which gives the runtime of the code
import time
def runner(func):
  def wrapper():
    start= time.time()
    func()
    print('The time taken by your code to run is', time.time()-start)
  return wrapper

@runner
def myfullname():
  time.sleep(5)
  print('Keshav Mahesh Joshi')


myfullname()
```

```python
# Let's try another example

import time

def runner(func):
    def wrapper():
        start= time.time()
        func()
        print('Your Runtime is', time.time()- start)

    return wrapper


@runner
def square(num):
    squared_num = num*num
    print(squared_num)

square(5)
```

**This error is due to, we are giving 1 positional argument thatis 'num' but in decorator the func does not receiving any argument, So according to this problem we have to make changes in our decorator.**

```python
# User Defined Runtime Decorator which works for almost all functions

import time

def runtime(func):
    def wrapper(*args, **kwargs):
        start= time.time()
        result= func(*args, **kwargs)
        print('Your Runtime is', time.time()- start)
        return result
    return wrapper

@runtime
def square(num):
    time.sleep(2)
    print(num*num)

# The simplest **kwargs function possible
@runtime
def show_info(**kwargs):
    time.sleep(5)
    print(kwargs)


square(26)

show_info(name= 'Keshav', gender= 'Male', age= '25')
```
